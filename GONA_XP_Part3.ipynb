# Introduction

In this tutorial we show how to use the Kalman filter for large scale regression and binary classification problems. 

In this part, we review classical stochastic optimization approach for logistic regression and show how to use variants of the linear Kalman filter to solve classification problem. Two variants are studied:

- The extended Kalman filter based on local second order approximation which is equivalent to the online Newton or online IRLS algorithm.
- The quadratic Kalman filter based on a global second order approximation which optimize a surrogate quadratic function of the logistic loss.
- The variational Kalman filter based on the iterative variational approximation developped in our paper.

# Logistic regression with Kalman filter

# Stochastic Optimization 

As before we begin with baseline algorithms to optimize logistic loss in an online way. We can apply directely the online Newton to the gradient and Hessian of the "one sample" logistic loss to obtain an online version of the IRLS algorithm. If we consider the one sample loss:

\begin{align}
l_i(x)&=-y_i\log(\sigma(u_i^Tx))-(1-y_i)\log(1-\sigma(u_i^Tx)), \\
&
\end{align}

the gradient and the Hessian are:

\begin{align}
\nabla_x l_i(x)&= -(y_i-\sigma(u_i^Tx))u_i\\
\nabla^2_x l_i(x)&= u_i \sigma(u_i^Tx)(1-\sigma(u_i^Tx)) u_i^T.\\
&
\end{align}

```python
import random
import numpy as np
import math
import numpy.linalg as LA
import matplotlib.pyplot as plt
```

```python
from LargeScaleKalmanLibrary.StochasticOptim import stochasticGradient, onlineNewton

def grad_loglossi(ui,yi,x):
    vi=ui.T.dot(x)
    vi=np.clip(vi,-50,50)
    sigmoid=1/(1+np.exp(-vi))
    return -(yi-sigmoid)*ui

def hessian_loglossi(ui,yi,x):
    eta=np.clip(ui.T.dot(x),-50,50)
    sigmoid=1/(1+np.exp(-eta))
    R=max(sigmoid*(1-sigmoid),1e-12)
    return np.outer(ui,ui)*R
```

We consider datas which are not perfectely separable with some noisy lables. Indeed the logistic loss is less sharp with some significant noise since the Hessian is proportional to $\sigma(1-\sigma)$.

# The extended Kalman filter for classification

We consider the non linear observations equations: $\hat{y}_t=\mu_t=h_t(x_t)=\sigma(u_t^Tx_t)$. One can remark the prediction of the observations $\hat{y}_t=\mu_t$ are continuous mean values whereas the true observations are discrete values $y_t=\{0,1\}$. The Kalman filter is used as a continuous regressor of the mean values of the output. To use the linear Kalman equations we need to linearize the function $h_t$ around the current estimate. 

## The iterative linearization process

Curious reader will find here some justification for the extended Kalman filter equations, this part can be skip to go directely to the short form of equations. Suppose we have a guess $x^*$ which is not far from the optimal (this guess is called reference trajectory in the Kalman dynamic framework) such that we can linearize around it the observation equation:

\begin{align*}
&h_t(x_t)-h_t(x^*)\approx h_t^\prime(x^*)(x_t-x^*)\\
&\delta y_t=H_t\delta x_t,\\
&\\
\end{align*}

where $H_t=h_t^\prime(x^*)$ is the observation matrix, $\delta x_t=x_t-x^*$ is the increment at step $t$ and $\delta y_t = y_t -h_t(x^*)$ is the incremental observation. If we consider $x^*$ is constant we have also the following relation:


\begin{align*}
&P_t=\mathrm{Cov}(x_t) = \mathrm{Cov}(\delta x_t),\\
&R_t=\mathrm{Cov}(y_t) = \mathrm{Cov}(\delta y_t),\\
&\\
\end{align*}

and we can use the linear Kalman equations to estimate the increment $\delta x_t$ in function of the incremental observations $\delta y_t$:


\begin{align*}
& \textbf{Linear Kalman filter}\\
&\textbf{around a reference trajectory} \\
&\delta x_t=\delta x_{t-1}+K_{t}(\delta y_t-H_t\delta x_{t-1}) \\
&K_t=P_{t-1}H_{t}^T(R_{t}+H_{t}P_{t-1}H_{t}^T)^{-1}\\
&P_{t}^{-1}=P_{t-1}^{-1}+ H_t^TR_t^{-1}H_t.\\
&
\end{align*}

Obviously we do not know the reference trajectory $x^*$. The principle of the extended Kalman filter is to take for this reference trajectory a recursive trajectory based on the last estimate $x_{t-1}$ and to consider this last estimate as a constant (not random). By doing that we do not manage temporal correlations of the random variable $x_t$ with time and the filter may diverge. If we consider $x^*=x_{t-1}$, we have now $\delta x_{t-1}=0$ and the updates simplify to:

\begin{align*}
&\delta x_t=\delta x_{t-1}+K_{t}(\delta y_t-H_t\delta x_{t-1})=K_{t}\delta y_t \\
&x_t=x_{t-1}+K_{t}(y_t-h_t(x_t)).
\end{align*}

$\delta y_t = y_t -h_t(x_{t-1})$ can be interpreted as the error between the current observation and the predicted one and is called the innovation error. We can add a model noise on the covariance matrix to correct the error made by the local approximation  $\mathrm{Cov}(x_t) \approx \mathrm{Cov}(\delta x_t)= \mathrm{Cov}(x_t-x_{t-1})$ and prevent divergence. In this case we have also to modify the update of $P_t$ as :

\begin{align*}
P_{t}^{-1}=(P_{t-1}+Q_t)^{-1}+ H_t^TR_t^{-1}H_t,\\
\end{align*}

where $Q_t$ is called covariance matrix of process (or model) noise.

$\textbf{Relation to other optimization algorithms:}$

As the Gauss-Newton algorithm, the extended Kalman filter for a stationary state is based on the linearization of a function inside a least square loss. There is however two main differences : 

- the linearization process is iterative for Kalman filter whereas it has a batch form for Gauss-Newton
- the least square loss is weighted with a covariance of noise $R_t$ which can change at each term. 

The extended Kalman filter is in fact more related to the online Newton descent. It is indeed strictely equivalent to online Newton descent for a specific class of function which belong to the exponential family.

## The Extended Kalman filter

As seen in the previous section, when the function which relate outputs $y$ to input $u$ is non linear, this function can be linearized around the last estimated $x_t$ such that we retrieve the linear Kalman equations but where the matrix $H$ depend now on $x_t$. When we do that for the sigmoid function $\mu_{t+1}=\sigma(u_{t+1}^Tx_t)=1/(1+e^{-u_{t+1}^Tx_t})$ we obtain the extended Kalman filter for logistic classification:

\begin{align}
R_{t+1}&=\mu_{t+1}(1-\mu_{t+1})=\sigma(u_{t+1}^Tx_t)(1-\sigma(u_{t+1}^Tx_t)) \\
H_{t+1}&=\nabla_x \sigma(u_{t+1}^Tx_t) = \sigma(u_{t+1}^Tx_t)(1-\sigma(u_{t+1}^Tx_t))u_{t+1}^T=R_{t+1}u_{t+1}^T\\
%P_{t+1}^{-1}&= P_t^{-1} + u_{t+1}R_{t+1}u_{t+1}^T  \\
x_{t+1} &=x_t + K_{t+1}(y_{t+1}-\sigma(u_{t+1}^Tx_t)) \\
K_{t+1}&=P_{t}u_{t+1}R_{t+1}(R_{t+1}+R_{t+1}u_{t+1}P_{t}u_{t+1}^TR_{t+1})^{-1} \\
P_{t+1}&=(\mathbb{I}-K_{t+1}R_{t+1}u_{t+1}^T)P_{t}\\
P_{t+1}^{-1}&=P_{t}^{-1} + u_{t+1}R_{t+1}u_{t+1}^T.
\end{align}

For Bernouilli variable, the variance $R_{t+1}$ can be computed from the past estimated $x_t$, we do not need anymore to specify it apriori. 

If we take a look more carefully at update equation using the relation $K_{t}=P_{t}H^T_{t}R^{-1}_{t}$:

\begin{align}
x_{t+1} &=x_t + K_{t+1}(y_{t+1}-\sigma(u_{t+1}^Tx_t)) \\
&=x_t + P_{t+1}H^T_{t+1}R^{-1}_{t+1}(y_{t+1}-\sigma(u_{t+1}^Tx_t)) \\
&=x_t + P_{t+1} u_{t+1}(y_{t+1}-\sigma(u_{t+1}^Tx_t)) \\
&=x_t - P_{t+1} \nabla l_i(x_{t}), \\
&
\end{align}

we find a gradient descent pre-conditionned by the matrix $P_{t+1}$ which is nothing else than the inverse of the Hessian matrix. We will check this correspondance experimentaly.

Finally the EKF updates can be rewritten in a simpler form when applied for logistic regression:

\begin{align}
&\textbf{ The extended Kalman filter for logistic regression}\\
&x_{t+1}=x_{t}+P_{t+1}u_{t+1}(y_{t+1}-\sigma(u_{t+1}^Tx_t)) \nonumber \\
& P_{t+1}^{-1}=P_{t}^{-1} + \sigma^\prime(u_{t+1}^Tx_{t}) u_{t+1}u_{t+1}^T. \nonumber
\end{align}

```python
from LargeScaleKalmanLibrary.MathUtils import sigmoid

def ExtendedKalmanFilter(x0,P0,inputs,outputs,passNumber=1,normalizeDatas=False):
    """
    ExtendedKalmanFilter (Logistic loss)

    Parameters
    ----------
    x0 : vector (d x 1)
        the initial guess for parameters. 
    P0 : matrix (d x d)
        the initial guess for covariance of parameters. 
    inputs : matrix N x md 
        the inputs in a matrix form: each row correspond to a 
        matrix H_t m x d put in vector form
    outputs : matrix N x m
        the outputs in na matrix form: each row correspond to a sample
    passNumber : int
        number of times we read the training datas (one pass by default)
    normalizeDatas : bool
        normalize inputs or not
    Returns
    -------
    xt : vector (d x 1)
        the estimated parameter. 
    Pt : matrix (d x d)
        the covariance on estimate.
    history_x : matrix N x d
        the history of all past estimates (results of iterations of algorithm).
    history_P : matrix N x dd
        the history of all past covariance estimates. 
    """
    
    d=x0.shape[0] #dimension of state
    N=outputs.shape[0] #number of samples
    m=outputs.shape[1] #dimension of outputs
    x=x0 
    P=P0
    nbIter=0 # number of iterations
    invP=LA.inv(P0)
    # we stock intermediate results in array where 
    # number of rows match number of observations
    history_x = np.zeros((N*passNumber+1,d))
    history_x[0,:]=x0.flatten()
    history_P = np.zeros((N*passNumber+1,d*d))
    history_P[0,:]=P0.flatten()
    history_e = np.zeros((N*passNumber+1,1))
    history_e[0]=0.5
    history_Pe = np.zeros((N*passNumber+1,1))
    history_Pe[0]=0.5
    
    if normalizeDatas:
        Unormed=LA.norm(np.std(inputs,axis=0))
        inputs=inputs/Unormed
          
    debug=False
    for numeroPass in range(1,passNumber+1):   
        for t in range(0,N): 
            
            # get data
            y=outputs[t].reshape(m,1)
            U=inputs[t].reshape(d,m)
            
            # compute R
            mu=sigmoid(U.T.dot(x)) 
            R=max(mu*(1-mu),1e-12)
            H=R*U.T

            # prediction error
            err=y-mu
        
            # update equations
            # computation of optimal gain
            S=R+H.dot(P).dot(H.T)
            K=P.dot(H.T).dot(LA.inv(S))
        
            # update state and covariance of state
            x=x+K.dot(err)
            P=P-K.dot(H).dot(P)
            #invP=invP+H.T.dot(LA.inv(R)).dot(H)
            #P=LA.inv(invP)
            if debug==True:
                print('x=',x)
                print('mu=',mu)
                print('S=',S)
                print('P=',P)
            
            history_x[nbIter+1,:]=x.flatten()
            history_P[nbIter+1,:]=P.flatten()
            history_e[nbIter+1]=err
            history_Pe[nbIter+1]=S
            
            nbIter=nbIter+1
        
        if numeroPass>1:
            # To manage different pass, shuffle the dataset
            DataSet=list(zip(inputs,outputs))
            random.shuffle(DataSet)
            inputs,outputs = zip(*DataSet)
            
        if normalizeDatas:
            ratio=1/Unormed 
            #x=x*ratio
            #P=P*ratio**2
            #history_x=history_x*ratio
            #history_P=history_P*ratio**2
        
    return x,P,history_x,history_P,history_e, history_Pe
```

```python
Nsamples=200
d=2
c=2
balance_shift=0
sigma=1
mu=0.7
epsilon=0.1
Y,U,x_opt,MU,Cov_u=LogLoss_data_generator(Nsamples,d,c,\
                                           balance_shift,epsilon,mu,sigma,\
                                          seed=2,covarianceInputs='isotropic')

x0=np.ones([d,1])*1e-3
numberPass=2
sigma0=1e6
Hessian0=np.identity(d)/sigma0**2
P0=LA.inv(Hessian0) #the initial covariance is given by the inverse of the initial Hessian 
x_on,P_on,history_x_on,history_P_on = onlineNewton(x0,Hessian0,U,Y,grad_loglossi,hessian_loglossi,numberPass)
x_ka,P_ka,history_x_ka,history_P_ka,history_e,history_Pe = ExtendedKalmanFilter(x0,P0,U,Y.reshape(-1,1),numberPass)
x_ada,P_ada,history_x_ada,history_P_ada = stochasticGradient(x0,U,Y,grad_loglossi,stepADAGRAD,numberPass,\
                                                         typeOptim='adagrad',sigma0=sigma0)
x_gn,P_gn,history_x_gn,history_P_gn = stochasticGradient(x0,U,Y,grad_loglossi,stepADAGRAD,numberPass,\
                                                         typeOptim='gauss-newton',sigma0=sigma0)

fig,(ax1,ax2,ax3,ax4)=plt.subplots(1,4,num=1,figsize=(20,3))
plotLogLoss(ax1,U,Y,x_opt,numberPass,[history_x_ka],['kalman filter'])
plotLogLoss(ax2,U,Y,x_opt,numberPass,[history_x_on],['online newton'])
plotLogLoss(ax3,U,Y,x_opt,numberPass,[history_x_ada],['adagrad'])
plotLogLoss(ax4,U,Y,x_opt,numberPass,[history_x_gn],['gauss-newton'])

plt.show()
```

As expected, the extended Kalman filter give exactely the same results than online Newton whereas Gauss-Newton approach failed. To make the Gauss-Newton converge we must rather use the generalized Gauss-Newton (see Martens).

# The quadratic approach

## Surrogate quadratic function on logistic loss (the $\eta$ trick)

The logistic loss have a sharp form particulary when $\sigma(u^Tx)$ is small. The quadratic  approach use the convexity of the logistic loss $l_i(x)$ with respect to $x^2$ to construct a quadratic surrogate function which is easier to optimize than the logistic loss. The surrogate quadratic function takes the following form (see [2] for more details):

The lower bound on the sigmoid function around a point $\xi$ writes : 

\begin{align*}
&\sigma(x)\geq \sigma(\xi)\exp(\frac{x-\xi}{2} + \eta (x^2-\xi^2))\\
&\text{with } \eta=-\frac{1}{2\xi}[\sigma(\xi)-\frac{1}{2}].\\
&
\end{align*}

We can then deduce a quadratic upper bound on the one sample loss $l_i(x)=-\log p(y_i/\mu_i=\sigma(u_i^Tx))$ around the point $\xi$:

\begin{align*}
&l_i(x,u_i,y_i) \leq   -  x^T \eta u_iu_i^Tx +(\frac{1}{2}-y_i)u_i^Tx + \frac{\xi}{2} + \eta \xi^2  - \log \sigma(\xi), \\
&
\end{align*}

[2] Bayesian parameter estimation via variational methods. Jaakkola & Jordan. 1998.

We illustrate on a particular point $\xi$ the bounds for the sigmoid and the one sample loss:

```python
def eta(xi):
    return -1/(2*xi)*(sigmoid(xi)-0.5)

def sigmoid_lower_bound(x,xi):
    a=(x-xi)/2 + eta(xi)*(x**2-xi**2)
    a=np.clip(a,-50,50)
    return sigmoid(xi)*np.exp(a)

def logloss(x,ui,yi):
    a=(2*yi-1)*ui*x
    a=np.clip(a,-50,50)
    return np.log(1+np.exp(-a))

def logloss_upper_bound(x,xi,ui,yi):
    H=-eta(xi)*ui**2
    v=(0.5-yi)*ui
    c=0.5*xi+eta(xi)*xi**2-math.log(sigmoid(xi))
    return H*x**2+v*x+c

fig,(ax1,ax2)=plt.subplots(1,2,figsize=(10,3))
t=np.arange(-10,10,0.1)
ax1.plot(t,sigmoid(t),label='sigmoid')
xi=3
ui=1
yi=1
ax1.plot(t,sigmoid_lower_bound(t,xi),label=r'lower-bound: $\xi$={}'.format(xi))
ax1.legend(loc='upper left')
ax1.axvline(xi,color='r',linestyle='-.') 
ax2.plot(t,logloss(t,ui,yi),label='logistic loss')
ax2.plot(t,logloss_upper_bound(t,xi,ui,yi),label=r'upper-bound: $\xi$={}'.format(xi))
ax2.axvline(xi,color='r',linestyle='-.') 
ax2.legend(loc='upper right')
plt.show()
```

## The batch variational logistic regression

We can write also the bound on the logistic loss where we consider different coefficents $\xi_i$ associated to each input $u_i$:

\begin{align*}
&l_t(x)=\sum_{i=1}^{t} l_i(x,u_i,y_i) \leq   - x^T (\sum_{i=1}^{t} \eta(\xi_i) u_iu_i^T) x +\sum_{i=1}^{t}(\frac{1}{2}-y_i)u_i^Tx + \sum_{i=1}^{t} \frac{\xi_i}{2} + \eta(\xi_i) \xi_i^2  - \log \sigma(\xi_i), \\
&
\end{align*}

Rather than minimize the "sharp" loss $l_t$, we can now minimize the quadratic surrogate function: the minimum can be expressed in function of the coefficient $\xi_i$ which in turn can be expressed in function of $x$ giving the following recursive coordinate descent equations: 

\begin{align}
&\textbf{ Batch variational logistic regression:} \\
&\text{ (iterate until convergence)}\\
&H=\sum_{i=1}^{t} - \eta(\xi_i)  u_iu_i^T\\
&v=\sum_{i=1}^{t}(\frac{1}{2}-y_i)u_i\\
& x= -\frac{1}{2}H^{-1}v \\
& \xi_i^2 = u_i^T(\frac{1}{2}H^{-1} +xx^T)u_i.\\
&
\end{align}

Let us test it on a 2D dataset:

```python
from LargeScaleKalmanLibrary.Evaluation import plotLogLossShape, plotLogLossUBShape

# Compute the quadratic form for a given xi
def QuadraticUB(xi,U,Y):
    H=-U.T.dot(eta(xi)*U)
    v=np.sum((0.5-Y)*U,axis=0)
    c=np.sum(0.5*xi+eta(xi)*xi**2-np.log(sigmoid(xi)))
    return H,v,c

# BATCH REGRESSION
def VariationalLogisticRegression(x0,U,Y):
    N,d=U.shape
    xi=np.ones([N,1])
    epsilon = 1e-6
    x=x0
    xNew=np.ones([d,1])*100
    history_xi=[]
    history_x=[]
    while LA.norm(xNew-x)>epsilon:
        x=xNew
        H,v,c=QuadraticUB(xi,U,Y)
        P=LA.inv(H)
        xNew=-0.5*P.dot(v)
        M=0.5*P+np.outer(xNew,xNew)
        xi=np.sqrt(np.sum(U.dot(M)*U,axis=1)).reshape(N,1)
        history_x.append(xNew)
        history_xi.append(xi)
    return xNew,xi,history_x,history_xi

Nsamples=200
d=2
c=2
balance_shift=0
sigmaLabel=1
mu=0.7
epsilon=0.3
Y,U,x_opt,MU,Cov_u=LogLoss_data_generator(Nsamples,d,c,\
                                           balance_shift,epsilon,mu,sigmaLabel,\
                                          seed=2,covarianceInputs='isotropic')
clf = LogisticRegression(penalty='none',random_state=0).fit(U, Y.reshape(Nsamples,))
batch_opt=clf.coef_.reshape(d,1)

x0=np.ones([d,1])*1e-3
x_vi,xi,history_x,history_xi=VariationalLogisticRegression(x0,U,Y)
print('the optimal find by batch variational logreg is: {}'.format(x_vi))
print('the loss find by batch variational logreg  is: {}'.format(LogLoss(Y,U,x_vi)))
print('the score find by batch variational logreg  is: {}'.format(LogScore(Y,U,x_vi)))

fig = plt.figure(3,figsize=(15,5))

ax = fig.add_subplot(121)
plotLogLossUBShape(ax,xi,batch_opt,U,Y,color='k',nbcontours=20,spread=50,Projection3D=False)
plotLogLossShape(ax,batch_opt,U,Y,color='k',nbcontours=20,spread=50,Projection3D=False)
ax.set_title(r'logLoss and his upper bound for $||\xi||={0:1.1f}$'.format(LA.norm(xi)))

ax = fig.add_subplot(122,projection='3d')
plotLogLossUBShape(ax,xi,batch_opt,U,Y,color='k',nbcontours=10,spread=20,Projection3D=True)
plotLogLossShape(ax,batch_opt,U,Y,color='k',nbcontours=10,spread=20,Projection3D=True)
ax.set_title(r'logLoss and his upper bound for $||\xi||={0:1.1f}$'.format(LA.norm(xi)))
#plt.savefig('fig2')
plt.show()
```

## The quadratic Kalman filter

The batch process can be done iteratively in a stochastic way: at each step the variational parameter $\xi_t$ associated to the new data $u_t$ is computed in function of the current estimate $x_t$. The Hessian, or information matrix P_t^{-1}, is construct in an iterative way as for the online Newton: $P_t^{-1}=P_{t-1}^{-1} -2 \eta(\xi_i)  u_iu_i^T$. Finally we can rewrite the equations to let appear an error in prediction $\varepsilon=R_{t+1}(y_{t+1}-\frac{1}{2})-u_{t+1}^Tx_t$ and a gain $K$ to form what we call the quadratic  Kalman filter:

\begin{align}
& \textbf{The quadratic Kalman filter}  \\
\xi_{t+1} &= \sqrt{u_{t+1}^T(P_{t} + \hat{x}_{t}\hat{x}_{t}^T)u_{t+1}}\\
R_{t+1}^{-1}&=\frac{1}{\xi_{t+1}}(\frac{1}{1+e^{-\xi_{t+1}}} -\frac{1}{2}) \\
%P_{t+1}^{-1}&=P_{t}^{-1}+u_{t+1}R_{t+1}u_{t+1}^T \\
\hat{x}_{t+1}&=\hat{x}_{t} + K_{t+1}(R_{t+1}(y_{t+1}-\frac{1}{2})-u_{t+1}^Tx_t))\\
K_{t+1}&=P_{t}u_{t+1}(R_{t+1}+u_{t+1}^TP_{t}u_{t+1})^{-1} \\
P_{t+1}&=(\mathbb{I}-K_{t+1}u_{t+1}^T)P_{t}\\
P_{t+1}^{-1}&=P_{t}^{-1} + u_{t+1}R_{t+1}^{-1}u_{t+1}^T.\\
\end{align}  

```python
def QuadraticKalmanFilter(x0,P0,inputs,outputs,passNumber=1,normalizeDatas=False):
    """
    QuadraticKalmanFilter 

    Parameters
    ----------
    x0 : vector (d x 1)
        the initial guess for parameters. 
    P0 : matrix (d x d)
        the initial guess for covariance of parameters. 
    inputs : matrix N x md 
        the inputs in a matrix form: each row correspond to a 
        matrix H_t m x d put in vector form
    outputs : matrix N x m
        the outputs in na matrix form: each row correspond to a sample
    passNumber : int
        number of times we read the training datas (one pass by default)
    normalizeDatas : Bool
        if true datas are normalized to 1
    Returns
    -------
    xt : vector (d x 1)
        the estimated parameter. 
    Pt : matrix (d x d)
        the covariance on estimate.
    history_x : matrix N x d
        the history of all past estimates (results of iterations of algorithm).
    history_P : matrix N x dd
        the history of all past covariance estimates. 
    """
    
    d=x0.shape[0] #dimension of state
    N=outputs.shape[0] #number of samples
    m=outputs.shape[1] #dimension of outputs
    x=x0 
    P=P0
    nbIter=0 # number of iterations
   
    # we stock intermediate results in array where 
    # number of rows match number of observations
    history_x = np.zeros((N*passNumber+1,d))
    history_x[0,:]=x0.flatten()
    history_P = np.zeros((N*passNumber+1,d*d))
    history_P[0,:]=P0.flatten()
    
    if normalizeDatas:
        Unormed=LA.norm(np.std(inputs,axis=0))
        inputs=inputs/Unormed
          
    debug=False
    for numeroPass in range(1,passNumber+1):   
        for t in range(0,N): 
            
            # get data
            y=outputs[t].reshape(m,1)
            U=inputs[t].reshape(d,m)            
                
            # compute matrix R
            xx=np.outer(x,x)
            xi=math.sqrt(U.T.dot(P+xx).dot(U))
            invR=np.ones([1,1])*(-2*eta(xi))
            R=LA.inv(invR)
            
            # compute gain K
            H=U.T
            S=R+H.dot(P).dot(H.T)
            K=P.dot(H.T).dot(LA.inv(S))
                            
            #update x
            x=x+K.dot(R.dot(y-0.5)-H.dot(x))
                
            #update P
            P=P-K.dot(H).dot(P)
            
            # update equations
            if debug==True:
                print(x)
            
            history_x[nbIter+1,:]=x.flatten()
            history_P[nbIter+1,:]=P.flatten()
            
            nbIter=nbIter+1
        
        if numeroPass>1:
            # To manage different pass, shuffle the dataset
            DataSet=list(zip(inputs,outputs))
            random.shuffle(DataSet)
            inputs,outputs = zip(*DataSet)
            
        if normalizeDatas:
            ratio=1/Unormed 
            #x=x*ratio
            #P=P*ratio**2
            #history_x=history_x*ratio
            #history_P=history_P*ratio**2
        
    return x,P,history_x,history_P
```

## The variational Kalman filter

The variational Kalman filter (or averaged online Newton) for logistic regression is defined by the following update equations where $\beta=\sqrt{\frac{8}{\pi}}$ (see our paper for more details):

\begin{align}
&\textbf{The variational Kalman filter for logistic regression} \label{logregIVGA}\\
& v_t= \frac{\beta}{\sqrt{x_t^TP_{t-1}x_t+\beta^2}} \nonumber \\
&P_t^{-1}=P_{t-1}^{-1}+v_t\sigma(v_t x_t^T\mu_{t-1} )(1-\sigma(v_t x_t^T\mu_{t-1} ))x_t x_t^T \nonumber \\
&P_{t}=P_{t-1}-P_{t-1}u_t(\frac{1}{v_t \sigma(v_t x_t^T\mu_{t-1})(1-\sigma(v_t x_t^T\mu_{t-1}))}+u_t^TP_{t-1}u_t)^{-1}u_t^T P_{t-1} \\
&\mu_t=\mu_{t-1} +P_{t} x_t  (t_t- \sigma (v_t x_t^T\mu_{t-1} )),  \nonumber 
\end{align}  

```python
from LargeScaleKalmanLibrary.MathUtils import sigmoid, cross

def VariationalKalmanFilterEXPLICIT(x0,P0,inputs,outputs,passNumber=1,normalizeDatas=False):
    """
    VariationalKalmanFilter (Logistic loss)

    Parameters
    ----------
    x0 : vector (d x 1)
        the initial guess for parameters. 
    P0 : matrix (d x d)
        the initial guess for covariance of parameters. 
    inputs : matrix N x md 
        the inputs in a matrix form: each row correspond to a 
        matrix H_t m x d put in vector form
    outputs : matrix N x m
        the outputs in na matrix form: each row correspond to a sample
    passNumber : int
        number of times we read the training datas (one pass by default)
    normalizeDatas : Bool
        if true datas are normalized to 1
    Returns
    -------
    xt : vector (d x 1)
        the estimated parameter. 
    Pt : matrix (d x d)
        the covariance on estimate.
    history_x : matrix N x d
        the history of all past estimates (results of iterations of algorithm).
    history_P : matrix N x dd
        the history of all past covariance estimates. 
    """
    beta=math.sqrt(8/math.pi)
    
    d=x0.shape[0] #dimension of state
    N=outputs.shape[0] #number of samples
    m=outputs.shape[1] #dimension of outputs
    x=x0 
    P=P0
    nbIter=0 # number of iterations
    invP=LA.inv(P0)
    # we stock intermediate results in array where 
    # number of rows match number of observations
    history_x = np.zeros((N*passNumber+1,d))
    history_x[0,:]=x0.flatten()
    history_P = np.zeros((N*passNumber+1,d*d))
    history_P[0,:]=P0.flatten()
    history_e = np.zeros((N*passNumber+1,1))
    history_e[0]=0
    history_Pe = np.zeros((N*passNumber+1,1))
    history_Pe[0]=0
    
    if normalizeDatas:
        Unormed=LA.norm(np.std(inputs,axis=0))
        inputs=inputs/Unormed
          
    debug=False
    for numeroPass in range(1,passNumber+1):   
        for t in range(0,N): 
            
            # get data
            y=outputs[t].reshape(m,1)
            U=inputs[t].reshape(d,m)
            
            # intermediate variables
            uPu=U.T.dot(P.dot(U))
            Pu=P.dot(U)
            
            # compute sigma(a)
            v=beta/math.sqrt(uPu+beta**2)
            a=U.T.dot(x)*v

            # update of covariance of state
            k=1/(v*sigmoid(a)*(1-sigmoid(a)))
            k=max(k,1e-12)
            P=P-cross(Pu)/(k+uPu)
        
            # update state
            x=x+Pu*(y-sigmoid(a))

            history_x[nbIter+1,:]=x.flatten()
            history_P[nbIter+1]=P.flatten()
            history_e[nbIter+1,:]=y-sigmoid(a)
            history_Pe[nbIter+1,:]=sigmoid(a)*(1-sigmoid(a))*(1-v)
            
            nbIter=nbIter+1
        
        if numeroPass>1:
            # To manage different pass, shuffle the dataset
            DataSet=list(zip(inputs,outputs))
            random.shuffle(DataSet)
            inputs,outputs = zip(*DataSet)
            
        if normalizeDatas:
            ratio=1/Unormed 
            #x=x*ratio
            #P=P*ratio**2
            #history_x=history_x*ratio
            #history_P=history_P*ratio**2
        
    return x,P,history_x,history_P,history_e, history_Pe
```

```python
from scipy import optimize

f = lambda x, a, b, c: x+a*sigmoid(b*x)+c

def VariationalKalmanFilter(x0,P0,inputs,outputs,passNumber=1,normalizeDatas=False):
    """
    VariationalKalmanFilter (Logistic loss)

    Parameters
    ----------
    x0 : vector (d x 1)
        the initial guess for parameters. 
    P0 : matrix (d x d)
        the initial guess for covariance of parameters. 
    inputs : matrix N x md 
        the inputs in a matrix form: each row correspond to a 
        matrix H_t m x d put in vector form
    outputs : matrix N x m
        the outputs in na matrix form: each row correspond to a sample
    passNumber : int
        number of times we read the training datas (one pass by default)
    normalizeDatas : Bool
        if true datas are normalized to 1
    Returns
    -------
    xt : vector (d x 1)
        the estimated parameter. 
    Pt : matrix (d x d)
        the covariance on estimate.
    history_x : matrix N x d
        the history of all past estimates (results of iterations of algorithm).
    history_P : matrix N x dd
        the history of all past covariance estimates. 
    """
    beta=math.sqrt(8/math.pi)
    
    d=x0.shape[0] #dimension of state
    N=outputs.shape[0] #number of samples
    m=outputs.shape[1] #dimension of outputs
    x=x0 
    P=P0
    nbIter=0 # number of iterations
    invP=LA.inv(P0)
    # we stock intermediate results in array where 
    # number of rows match number of observations
    history_x = np.zeros((N*passNumber+1,d))
    history_x[0,:]=x0.flatten()
    history_P = np.zeros((N*passNumber+1,d*d))
    history_P[0,:]=P0.flatten()
    history_e = np.zeros((N*passNumber+1,1))
    history_e[0]=0
    history_Pe = np.zeros((N*passNumber+1,1))
    history_Pe[0]=0
    
    if normalizeDatas:
        Unormed=LA.norm(np.std(inputs,axis=0))
        inputs=inputs/Unormed
          
    debug=False
    for numeroPass in range(1,passNumber+1):   
        for t in range(0,N): 
            
            # get data
            y=outputs[t].reshape(m,1)
            U=inputs[t].reshape(d,m)
            
            # intermediate variables
            uPu=U.T.dot(P.dot(U))
            Pu=P.dot(U)
            
            # compute a
            v=beta/math.sqrt(uPu+beta**2)
            c=-U.T.dot(x) - uPu*y
            a=optimize.newton(f, 0, args=(uPu,v,c,))
            
            #updates
            x=x+Pu*(y-sigmoid(v*a))
            k=1/(v*sigmoid(v*a)*(1-sigmoid(v*a)))
            k=max(k,1e-12)
            #P=LA.inv(LA.inv(P)+v*sigmoid(v*a)*(1-sigmoid(v*a))*cross(U))
            P=P-cross(Pu)/(k+uPu)

            # save history
            history_x[nbIter+1,:]=x.flatten()
            history_P[nbIter+1]=P.flatten()
            history_e[nbIter+1,:]=y-sigmoid(v*a)
            history_Pe[nbIter+1,:]=sigmoid(v*a)*(1-sigmoid(v*a))*(1-v)
            
            nbIter=nbIter+1
        
        if numeroPass>1:
            # To manage different pass, shuffle the dataset
            DataSet=list(zip(inputs,outputs))
            random.shuffle(DataSet)
            inputs,outputs = zip(*DataSet)
            
        if normalizeDatas:
            ratio=1/Unormed 
            #x=x*ratio
            #P=P*ratio**2
            #history_x=history_x*ratio
            #history_P=history_P*ratio**2
        
    return x,P,history_x,history_P,history_e, history_Pe
```
